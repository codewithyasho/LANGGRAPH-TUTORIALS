```python
from langchain_community.tools import YouTubeSearchTool
from langgraph.checkpoint.memory import MemorySaver
from langchain.agents import create_agent
from langchain_ollama import ChatOllama
from pydantic import BaseModel, Field
from langchain_groq import ChatGroq
from dotenv import load_dotenv
import os

load_dotenv()
memory = MemorySaver()
```

```python
# llm schema
class yt_schema(BaseModel):
    yt_query: str = Field(
        description="youtube video topic, channel name, or video title to search for")
```

```python
llm = ChatGroq(model="llama-3.3-70b-versatile")
llm_with_schema = llm.with_structured_output(yt_schema)

user_query = input("What do you want to search for on youtube? ")

response = llm_with_schema.invoke(f"""
you will be given a query to search on youtube.
You need to extract the query and return it in a structured format that follows the schema provided.
you show resturn only the youtube video topic, channel name, or video title to search for in the format specified by the schema. 
Do not include any other information or explanation, just return the structured data.\n
The user query is: {user_query}
""")
```

```python
yt_search_query = response.yt_query
lst = yt_search_query.split(",")
print(lst)

result = " ".join(s.strip() for s in lst)
print(result)
```

```python
tool = YouTubeSearchTool()

yt_list = tool.run(f"{result}, 5")
```

```python
print(f"Here are the top YouTube videos: {yt_list}")
```
